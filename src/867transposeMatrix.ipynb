```
def transpose(A):
    row, colume =len(A),len(A[0])
    ans=[[None]*row for _ in range(colume)]
    for i, row in enumerate(A):
          for j, value in enumerate(row):
              ans[j][i]=value
    return ans
```

```
B=[[1,2,3],[6,7,8]]
print(transpose(B))
```
